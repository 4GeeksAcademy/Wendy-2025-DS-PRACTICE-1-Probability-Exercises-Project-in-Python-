## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

def probabilidad ():

    cuenta_de_lanzamientos_que_cumplen_metricas = 0

    for lanzamientos in range (1,1001):

        posibles_resultados_dado1=random.randint(1,7)
        posibles_resultados_dado2=random.randint(1,7)

        puntuaciones=posibles_resultados_dado1+posibles_resultados_dado2
        
        print(puntuaciones)

        if (puntuaciones > 7) or (puntuaciones % 2 == 0):
    
            cuenta_de_lanzamientos_que_cumplen_metricas += 1

    return cuenta_de_lanzamientos_que_cumplen_metricas/1000

print("La probabilidad de que la puntuación sea mayor a 7 o par, es de:",probabilidad()*100,"%")


```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random

caja_de_bolas = {}

# Crea la cajas con las bolas
for i in range(1,61):
    if i < 11:
        caja_de_bolas[i] = "White"
    elif (i > 10) and (i < 31):
        caja_de_bolas[i] = "Red"
    else:
        caja_de_bolas[i] = "Green"

lista_de_colores_en_caja=list(caja_de_bolas.values())

def probabilidad_1 ():
    
    cuenta_de_veces_que_se_sacan_bolas =0
   
    for veces_sacadas in range (1,1001):
        
        sacar_5_bolas=random.sample(lista_de_colores_en_caja,k=5)
        
        print(sacar_5_bolas)
        
        bolas_blancas=sacar_5_bolas.count("White")
        bolas_rojas=sacar_5_bolas.count("Red")

        if bolas_blancas == 3 and bolas_rojas == 2:
            cuenta_de_veces_que_se_sacan_bolas += 1
       
    return(cuenta_de_veces_que_se_sacan_bolas/1000)

porcentaje_de_probabilidad_1=probabilidad_1()*100



def probabilidad_2 ():
    
    cuenta_de_veces_que_se_sacan_bolas_iguales =0
   
    for veces_sacadas_iguales in range (1,1001):
        
        sacar_5_bolas_iguales=random.sample(lista_de_colores_en_caja,k=5)
        
        print(sacar_5_bolas_iguales)
        
        bolas_blancas=sacar_5_bolas_iguales.count("White")
        bolas_rojas=sacar_5_bolas_iguales.count("Red")
        bolas_verdes=sacar_5_bolas_iguales.count("Green")

        if bolas_blancas == 5 or bolas_rojas == 5 or bolas_verdes == 5:
            cuenta_de_veces_que_se_sacan_bolas_iguales += 1
       
    return(cuenta_de_veces_que_se_sacan_bolas_iguales/1000)

porcentaje_de_probabilidad_2=probabilidad_2()*100


print(f"La probabilidad de que salgan 3 bolas blancas y 2 rojas, es de:{porcentaje_de_probabilidad_1}%\nLa probabilidad de que salgan todas del mismo color, es de:{porcentaje_de_probabilidad_2}%") 
           
```
